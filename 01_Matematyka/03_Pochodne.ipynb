# Pochodna funkcji

## Pochodna funkcji jednej zmiennej w punkcie
Pochodna funkcji $f(x)$ w punkcie $x_0$ to granica ilorazu różnicowego:
$$
f'(x_0) = \lim_{h \to 0} \frac{f(x_0 + h) - f(x_0)}{h}
$$

## Pochodne cząstkowe

Dla funkcji wielu zmiennych $f(x_1, x_2, \ldots, x_n)$ pochodna cząstkowa po zmiennej $x_i$ to granica ilorazu różnicowego:

$$
\frac{\partial f}{\partial x_i} = \lim_{h \to 0} \frac{f(x_1, \ldots, x_i + h, \ldots, x_n) - f(x_1, \ldots, x_i, \ldots, x_n)}{h}
$$



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# przykłady pochodnych funkcji jednej zmiennej
x = sp.symbols('x')
# general function
f = sp.Function('f')(x)
g = sp.Function('g')(x)
```

```python
# pochodna sumy funkcji:  (f+g)' = f' + g' 
sp.diff(f+g, x)
```

```python
# pochodna iloczynu funkcji: (f*g)' = f'*g + f*g'
sp.diff(f*g, x)
```

```python
# pochodna ilorazu funkcji: (f/g)' = (f'*g - f*g')/g^2
sp.diff(f/g, x).simplify()
```

```python
# przykład funkcji sin(x)+cos(x)
z = sp.sin(x)
zdiff = sp.diff(z, x)
```

```python
# wykres funkcji i jej pochodnej
z_ = sp.lambdify(x, z, 'numpy')
zdiff_ = sp.lambdify(x, zdiff, 'numpy')
x_ = np.linspace(-2*np.pi, 2*np.pi, 100)
plt.plot(x_, z_(x_), label=f'f(x)={z}')
plt.plot(x_, zdiff_(x_), label=f"f'(x)={zdiff}")
plt.legend()
plt.show()
```

```python
# pochodne wielokrotne
for n in range(1, 10):
    z = sp.sin(x)
    for i in range(n):
        z = sp.diff(z, x)
    z_ = sp.lambdify(x, z, 'numpy')
    print(f'f^{n}(x)={z}')
```

```python
# pochodne wielokrotne dla sin(cos(x))
z = sp.sin(sp.cos(x))
for n in range(1, 4):
    z = sp.diff(z, x)
    z_ = sp.lambdify(x, z, 'numpy')
    print(f'f^{n}(x)={z}')
```

# Pochodna funkcji wielu zmiennych


```python
# przykłady pochodnych funkcji wielu zmiennych
x, y = sp.symbols('x y')
# general function
f = sp.Function('f')(x, y)
g = sp.Function('g')(x, y)
```

```python
# pochodna sumy funkcji: (f+g)' = f' + g'
sp.diff(f+g, x)
```

# kolejne własności pochodnych funkcji wielu zmiennych są analogiczne do funkcji jednej zmiennej

```python
# policz pochodne cząstkowe funkcji sin(x)*exp(y)
f = sp.sin(x)*sp.exp(y)
f_x = sp.diff(f, x)
f_x
```

```python
f_y = sp.diff(f, y)
f_y
```

```python
# pochodna mieszana
f_xy = sp.diff(f, x, y)
f_xy
```
